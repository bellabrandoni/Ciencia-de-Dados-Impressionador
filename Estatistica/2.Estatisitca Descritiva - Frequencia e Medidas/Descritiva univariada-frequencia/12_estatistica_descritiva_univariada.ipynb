# Estatística descritiva univariada

A estatística descritiva descreve e sintetiza as principais características observadas em um conjunto de dados por meio de tabelas, gráficos e medidas-resumo. A análise é baseada no conjunto de dados disponível, sem inferir ou generalizar para uma população maior.

A estatística descritiva pode ser univariada, bivariada ou multivariada. A estatística descritiva univariada analisa uma única variável por vez, enquanto a bivariada analisa duas variáveis simultaneamente. A multivariada analisa três ou mais variáveis simultaneamente.

O tipo de variável é importante no cálculo de estatísticas descritivas e na escolha de gráficos.
